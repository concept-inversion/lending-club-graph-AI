# Separate columns into csv files

Transaction Details

```python
x=pd.DataFrame(df[["transaction_id","date_of_transaction","hour","installment","payment_amount","BeaAdmin"]],
             columns=["transaction_id","date_of_transaction","hour","installment","payment_amount","BeaAdmin"]
            ).reset_index().drop('index', 1).shift()[1:].to_csv(write_folder+"//transaction_details.csv",
                                                                index_label="transactionId:ID(Transaction)",
                                                                float_format="%.10g")
```

Staff Ids

```python
pd.DataFrame(df["staff_id"].unique(),columns=["staff_id"]).drop_duplicates(subset="staff_id").reset_index().drop('index', 1).shift()[1:].to_csv(write_folder+"//staff_ids.csv",index=True,index_label="staffId:ID(Staff)",float_format="%.10g")
```

Customer Info

```python
pd.DataFrame(df[["cust_id","name"]],
             columns=["cust_id","name"]).drop_duplicates(subset="cust_id").reset_index().drop('index', 1).shift()[1:].to_csv(write_folder+"//customer.csv",index=True,index_label = "customerId:ID(Customer)")
```

Vehicle info

```python
pd.DataFrame(df[["plate_no"]],
             columns=["plate_no"]).drop_duplicates(subset="plate_no").reset_index().drop('index', 1).shift()[1:].to_csv(write_folder+"//vehicles.csv",index=True,index_label = "vehicleId:ID(Vehicle)",float_format="%.10g")
```

Branch ID

```python
pd.DataFrame(df["branch_id"].unique(),columns=["branch_id"]).drop_duplicates(subset="branch_id").reset_index().drop('index', 1).shift()[1:].to_csv(write_folder+"//branches.csv",index=True,index_label = "branchId:ID(Branch)",float_format="%.10g")
```

Counter ids

```python
pd.DataFrame(df["counter_id"].unique(),columns=["counter_id"]).drop_duplicates(subset="counter_id").reset_index().drop('index', 1).shift()[1:].to_csv(write_folder+"//counters.csv",index=True,index_label = "counterId:ID(Counter)",float_format="%.10g")
```

Main table with the foreign keys

```python
df2 = pd.DataFrame(df[["transaction_id","staff_id","cust_id","plate_no","branch_id","counter_id"]],
             columns=["transaction_id","staff_id","cust_id","plate_no","branch_id","counter_id"])
```

```python
df2
```

We replace every column's values in df2 with the corresponding index from the appropriate table

```python
# Get all the filenames of the individual CSV files
for filename in os.listdir(write_folder):
    if filename==write_folder+".csv" or filename=="main_table.csv":
        continue
    print(filename)
    # Read the CSV file
    df3 = pd.read_csv(write_folder+"//"+filename)
    # Get the list of columns
    columns = list(df3)
    # Create a Dict to map the column values to an index
    dict3 ={}
    for index, row in df3.iterrows():
        dict3[row[columns[1]]] = row[columns[0]]
    # Map the column values to the dict
    df2[columns[1]] = df2[columns[1]].map(dict3)
```

```python
df2
```

```python
df2 = df2.dropna().reset_index(drop=True)
```
